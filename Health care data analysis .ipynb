# visualize the percentage of pepole getting govt helath insurence due to how income , due to old age also the percentage of pepole having private health insurance

```python
# govt
label = ['yes','no']
y = df[df['freepoor']=='yes']
n = df[df['freepoor']=='no']
x = [y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title(" % of pepole getting govt health insurance due to low income")
#private
label = ['yes','no']
y = df[df['private']=='yes']
n = df[df['private']=='no']
x = [y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title(" % of pepole getting private health insurance ")
#old age Govt
label = ['yes','no']
y = df[df['freerepat']=='yes']
n = df[df['freerepat']=='no']
x = [y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title(" % of pepole getting govt health insurance due to old age")
```

# plot a horizontal bar chart to Aanlyze the reduced days of activity due to illness bases on gender

```python
db = df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.bar(db['gender'],db['reduced'],color=['green','blue'])
plt.title('bar chart')
plt.xlabel('gender')
plt.ylabel('reduced acitivity')
plt.show()
```
